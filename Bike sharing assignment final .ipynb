```python
# Calculating VIF
vif=pd.DataFrame()
vif['Features']=df_train_rfe_x.columns
vif['VIF']=[variance_inflation_factor(df_train_rfe_x.values,i) for i in range(df_train_rfe_x.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
